# Exercise 5: NumPy Array Manipulation

```python
import numpy as np

# Matrix Creation: Create a 5Ã—5 matrix with values from 1 to 25
matrix = np.arange(1, 26).reshape(5, 5)
print("Generated 5x5 Matrix:")
print(matrix)
```

```python
# Diagonal Extraction: Extract the diagonal elements
diagonal = np.diag(matrix)
print("\nExtracted Diagonal Elements:")
print(diagonal)
```

```python
# Calculation: Calculate the sum of the diagonal elements
diagonal_sum = diagonal.sum()
print(f"\nSum of Diagonal Elements: {diagonal_sum}")
```
