# Exercise 6: Pandas Series Creation and Indexing

```python
import pandas as pd

# Part A: Create a Pandas Series from a list of temperatures in Celsius
celsius_temps = [22, 25, 19, 30, 27, 21, 18]
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
celsius_series = pd.Series(celsius_temps, index=days, name="Celsius")

print("Temperatures in Celsius:", celsius_series)
```

```python
# Part B: Convert temperatures to Fahrenheit and add as a new Series
fahrenheit_series = (celsius_series * 9/5) + 32
fahrenheit_series.name = "Fahrenheit"

print("\nTemperatures in Fahrenheit:", fahrenheit_series)
```

```python
comparison_df = pd.DataFrame({'Celsius': celsius_series, 'Fahrenheit': fahrenheit_series})
print(comparison_df)
```
